```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('loadresults/load_stats_history.csv')
df.dropna(subset=['Type'], inplace=True)
start_timestamp = df['Timestamp'].iloc[0]
df['Timestamp'] = df['Timestamp'].apply(lambda x: x - start_timestamp)
df.plot(x='Timestamp', y='User Count')
```

```python
results = df[['User Count']].to_numpy()
if results.shape[0] < 60:
    results = np.append(np.zeros(60-results.shape[0]), results)
print(results.shape)
results
```
